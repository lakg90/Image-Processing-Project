### Question 2

Further process the image, aiming to replace the light and dark blue tiles (second and fourth from the left) with white pixels, and to replace all other parts of the image with black pixels.

Make sure to process the *result* obtained from Question 1, rather than the original image.

Describe any decisions you make, and the methods used, in your LaTeX write-up. Comment on your results, including at least one relevant figure.

````python
# Add code here
img_f_1 = ip.median_filter(img, (1, 7), mode="reflect")

ip.imshow(img_f_1)

#changing the image to grey scale:

alpha = -50  # Contribution of red
beta = -50  # Contribution of green
gamma = 100 # Contribution of blue

img_gs = alpha * img_f_1[:, :, 0] + beta * img_f_1[:, :, 1] + gamma * img_f_1[:, :, 2]

ip.imshow(img_gs)

#thresholding the image

img_gs2 = 255 * (img_gs > 1000)

ip.imshow(img_gs2)
````
